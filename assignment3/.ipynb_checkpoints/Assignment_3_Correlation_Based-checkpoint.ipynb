```python
#Author: Sams Khan
#Clsss: BioInformatics

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from numpy.linalg import inv
import pandas as pd

#Reading in files
gene_data = pd.read_csv("Gene_expression.csv", header=None)
gr_truth= pd.read_csv("Adj_1.csv", header=None)


```

```python
#Preprocessing data
corr_matrix=gene_data.corr(method='pearson')**2
corr_matrix
```
